```python
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_addons as tfa
from tensorflow import keras
from tensorflow.keras import Model
from tensorflow.keras.layers import (
    Activation,
    BatchNormalization,
    Concatenate,
    Conv2D,
    Conv2DTranspose,
    MaxPooling2D,
    Input,
)

IMAGE_HEIGHT = 28
IMAGE_WIDTH = 28
IMAGE_DEPTH = 1
BATCH_SIZE = 256
EPOCHS = 20
NOISE = (0.1, 0.3)
CUT_OUT_SQUARE = 12


def activation(x):
    y = BatchNormalization()(x)
    y = Activation("relu")(y)
    return y


def conv_block(x, num_filters):
    x = Conv2D(num_filters, 3, padding="same")(x)
    x = activation(x)

    return x


def encoder_block(x, num_filters):
    x = conv_block(x, num_filters)
    p = MaxPooling2D((2, 2))(x)
    return x, p


def decoder_block(x, skip_features, num_filters):
    x = Conv2DTranspose(num_filters, (2, 2), strides=2, padding="same")(x)
    x = Concatenate()([x, skip_features])
    x = conv_block(x, num_filters)
    return x


def unet_model():
    fac = 4
    inputs = Input([IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH])

    s1, p1 = encoder_block(inputs, 64 / fac)
    p1 = Conv2D(64 / fac, 7)(p1)
    s2, p2 = encoder_block(p1, 128 / fac)
    s3, p3 = encoder_block(p2, 128 / fac)
    s4, p4 = encoder_block(p3, 128 / fac)

    b1 = conv_block(p4, 256 / fac)

    d1 = decoder_block(b1, s4, 128 / fac)
    d2 = decoder_block(d1, s3, 128 / fac)
    d3 = decoder_block(d2, s2, 128 / fac)
    d3 = Conv2DTranspose(64 / fac, 7)(d3)
    d4 = decoder_block(d3, s1, 64 / fac)

    outputs = Conv2D(IMAGE_DEPTH, 1, padding="same", activation="sigmoid")(d4)

    model = Model(inputs, outputs, name="U-Net")
    return model

@tf.function
def normalize(image, label):
    image = tf.cast(image, tf.dtypes.float32)
    image = tf.divide(image, 255.0)
    
    label = tf.cast(label, tf.dtypes.float32)
    label = tf.divide(label, 255.0)
    
    image = tf.reshape(image, (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH))
    label = tf.reshape(label, (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH))

    return image, label

@tf.function
def random_cut_out(images, labels):
    coin = tf.random.uniform((), 0.0, 1.0)
    if coin >= 0.5:
        images = tfa.image.random_cutout(images, CUT_OUT_SQUARE, constant_values=0)
    return images, labels

@tf.function
def gn(image, label):
    noise = tf.random.normal(shape=tf.shape(image), mean=0.0, stddev=tf.random.uniform((), *NOISE)
    , dtype=tf.float32)
    noise_img = tf.add(image, noise)    
    return noise_img, label

(x_train, _), (x_test, _) = keras.datasets.mnist.load_data()

train = tf.data.Dataset.from_tensor_slices((x_train, x_train)).map(normalize).repeat(EPOCHS).map(gn).shuffle(100).batch(BATCH_SIZE).map(random_cut_out)
test = tf.data.Dataset.from_tensor_slices((x_test, x_test)).map(normalize).map(gn).shuffle(100).batch(BATCH_SIZE).map(random_cut_out)

model = unet_model()

opt = tf.keras.optimizers.Adam(learning_rate=1e-3)
loss = tf.keras.losses.MeanSquaredError()
metrics = ["accuracy"]

model.compile(
    optimizer=opt,
    loss=loss,
    metrics=metrics,
)

tf.keras.utils.plot_model(model, show_shapes=True)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss", factor=0.5, patience=7, min_lr=1e-5
)

STEPS_PER_EPOCH = len(x_train) // BATCH_SIZE
VALIDATION_STEPS = len(test) // BATCH_SIZE
model.summary()

def show_predictions():
    maxx = 0
    for image, label in test:
        if maxx == 3:
            break
        maxx += 1
        den = model.predict(image)
        
        plt.figure(figsize=(15, 15))
        
        plt.subplot(1, 3, 1)
        plt.title("image")
        plt.imshow(image[0])
        plt.axis("off")
        
        plt.subplot(1, 3, 2)
        plt.title("label")
        plt.imshow(label[0])
        plt.axis("off")
        
        plt.subplot(1, 3, 3)
        plt.title("witouht noise")
        plt.imshow(den[0])
        plt.axis("off")
        
        plt.show()


class LossAndErrorPrintingCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs=None):
    show_predictions()

show_predictions()
history = model.fit(
    train,
    epochs=EPOCHS,
    steps_per_epoch=STEPS_PER_EPOCH,
    validation_data=test,
    callbacks=[reduce_lr, LossAndErrorPrintingCallback()],
)
```
